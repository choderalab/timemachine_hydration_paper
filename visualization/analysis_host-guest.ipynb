```python
def write(xyz, masses):
    buf = str(len(masses)) + '\n'
    buf += 'timemachine\n'
    for m, (x,y,z) in zip(masses, xyz):
        if int(round(m)) == 12:
            symbol = 'C'
        elif int(round(m)) == 16:
            symbol = 'O'
        elif int(round(m)) == 14:
            symbol = 'N'
        elif int(round(m)) == 1:
            symbol = 'H'
        else:
            raise Exception("Unknown mass:" + str(m))

        buf += symbol + ' ' + str(round(x,5)) + ' ' + str(round(y,5)) + ' ' +str(round(z,5)) + '\n'
    return buf
```

```python
def run_simulation(
    potentials,
    params,
    param_groups,
    conf,
    masses,
    dp_idxs):

    conf = conf
    
    potentials = forcefield.merge_potentials(potentials)
        
    dt = 0.0005
    ca, cb, cc = langevin_coefficients(
        temperature=298.15,
        dt=dt,
        friction=75,
        masses=masses
    )

    m_dt, m_ca, m_cb, m_cc = dt, 0.25, cb, np.zeros_like(masses)

    opt = custom_ops.LangevinOptimizer_f64(
        m_dt,
        m_ca,
        m_cb,
        m_cc
    )

    v0 = np.zeros_like(conf)
    dp_idxs = dp_idxs.astype(np.int32)

    ctxt = custom_ops.Context_f64(
        potentials,
        opt,
        params,
        conf, # x0
        v0, # v0
        # np.arange(len(params))
        dp_idxs
    )

    xyz_buffer = ''
    xyz_buffer = write(ctxt.get_x()*10,masses)
    # minimize the system
    for i in tqdm(range(10000)):
        ctxt.step()
#         if i % 75== 0:
#             buf = write(ctxt.get_x()*10, masses)
#             xyz_buffer += buf
#             print(ctxt.get_E())
#             print(ctxt.get_dx_dp())
    
    dt = 0.001
    opt.set_dt(dt)
    opt.set_coeff_a(ca)
    opt.set_coeff_b(cb)
    opt.set_coeff_c(cc)
            
    for i in tqdm(range(10000)):
        if i % 50 == 0:
            buf = write(ctxt.get_x()*10, masses)
            xyz_buffer += buf
        ctxt.step()
    
    return xyz_buffer
```

```python
dp_idxs = np.argwhere(guest_param_groups == 7).reshape(-1)
buf = run_simulation(combined_potentials, combined_params, combined_param_groups, combined_conf, combined_masses, dp_idxs=dp_idxs)
```
